```python
import numpy as np
import datetime
import plotly as py
import plotly.graph_objs as go
py.offline.init_notebook_mode(connected = True)
```

```python
path = "2019-05-07-04-01-09-854_NN_output.txt"
```

```python
logs = [[],[]]

with open(path, 'r') as f:
    for line in f:
        log = line.split()
        if len(log)==2:
            logs[0].append(datetime.datetime.strptime(log[0], "%Y-%m-%d-%H-%M-%S-%f"))
            logs[1].append(float(log[1]))
```

```python
NNProbTrace = go.Scatter(
    x = logs[0],
    y = logs[1],
    mode = 'lines+markers',
    name = 'speedMarkerY'
)

NNOutputTrace = go.Scatter(
    x = logs[0],
    y = [round(prob) for prob in logs[1]],
    mode = 'lines+markers',
    name = 'speedMarkerY'
)

#data = [accelYTrace]
data = [NNProbTrace, NNOutputTrace]
fig = go.Figure(data = data)

py.offline.iplot(fig, filename = "")
```
